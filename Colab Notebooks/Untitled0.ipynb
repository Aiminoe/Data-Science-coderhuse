```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def plot_ecg(data):
    # Personalizar el estilo de la gráfica con Seaborn
    sns.set(style="whitegrid", rc={"axes.facecolor": 'whitesmoke'})  # Establecer un fondo blanco

    plt.figure(figsize=(10, 6))
    plt.title('Electrocardiograma')

    # Graficar los datos utilizando Matplotlib y Seaborn
    plt.plot(data, color='b', label='ECG Data')

    # Personalizar la gráfica para ECG
    plt.xlabel('Tiempo (ms)')
    plt.ylabel('Amplitud (mV)')
    plt.legend(loc='best')

    # Mostrar la gráfica con una grilla punteada
    plt.grid(True, linestyle='--', linewidth=0.7, color='gray')

    plt.show()

# Ejemplo de uso
ecg_data = np.random.randn(100)  # Reemplaza con tus propios datos del ECG
plot_ecg(ecg_data)

```
