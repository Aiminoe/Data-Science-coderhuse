```python
import requests
import pandas as pd

# Obtener datos de usuarios
response = requests.get('https://jsonplaceholder.typicode.com/users')
users_data = response.json()

# Crear DataFrame
df = pd.DataFrame(users_data)

# Mostrar el DataFrame
df
```
